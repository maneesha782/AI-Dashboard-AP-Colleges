```python
total_colleges = df['College_Name'].nunique()

total_districts = df['District'].nunique()

avg_fees = df['Fees_Per_Year'].mean()

avg_placement = df['Placement_Percentage'].mean()

print("Total Colleges:", total_colleges)
print("Total Districts:", total_districts)
print("Average Fees:", avg_fees)
print("Average Placement:", avg_placement)
```

```python
import pandas as pd

df = pd.read_csv("AP_Engineering_Colleges.csv")
```

```python
import matplotlib.pyplot as plt

district_counts = df['District'].value_counts()


district_counts.plot(kind='bar')

plt.title('Number of Engineering Colleges by District')
plt.xlabel('District')
plt.ylabel('Number of Colleges')

plt.show()
```

Visakhapatnam and Guntur have more engineering colleges.
Some districts have fewer colleges.

```python
import matplotlib.pyplot as plt

df["Type"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Government vs Private Colleges")

plt.show()
```

Private colleges are more than government colleges.

```python
import matplotlib.pyplot as plt

branch_counts=df["Branch "].value_counts()
branch_counts.plot(kind="bar")
plt.title("Branch Distribution")
plt.xlabel("Branch")
plt.ylabel("Count")

plt.show()
```

CSE is one of the most commonly offered branches.

```python
plt.hist(df['Fees_Per_Year'], bins=5)

plt.title('Distribution of College Fees')
plt.xlabel('Fees Per Year')
plt.ylabel('Number of Colleges')

plt.show()
```

This graph shows how college fees are distributed.

```python


plt.scatter(df['Fees_Per_Year'],
            df['Placement_Percentage'])

plt.title('Fees vs Placement Percentage')
plt.xlabel('Fees Per Year')
plt.ylabel('Placement Percentage')

plt.show()
```

This graph shows whether higher fees are associated with better placements.
